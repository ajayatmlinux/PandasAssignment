```python
# First DataFrame
df1 = pd.DataFrame({'id': ['A01', 'A02', 'A03', 'A04'],
                    'Name': ['ABC', 'PQR', 'DEF', 'GHI']})
  
# Second DataFrame
df2 = pd.DataFrame({'id': ['B05', 'B06', 'B07', 'B08'],
                    'Name': ['XYZ', 'TUV', 'MNO', 'JKL']})
  
  
frames = [df1, df2]
  
result = pd.concat(frames)
display(result)
```

```python
frames
```

```python
result = pd.concat(frames, axis=1)
display(result)
```

```python
result = pd.concat(frames, join='inner')
display(result)
```

```python
result = pd.concat([df1,df2])
display(result)
```

```python
# creating a dataframe
df1 = pd.DataFrame({'Name':['Raju', 'Rani', 'Geeta', 'Sita', 'Sohit'],
                    'Marks':[80, 90, 75, 88, 59]})
  
# creating another dataframe with different data
df2 = pd.DataFrame({'Name':['Raju', 'Divya', 'Geeta', 'Sita'],
                    'Grade':['A', 'A', 'B', 'A'],
                    'Rank':[3, 1, 4, 2 ],
                    'Gender':['Male', 'Female', 'Female', 'Female']})
# display df1
display(df1)
  
# display df2
display(df2)
```

```python
df2.merge(df1)
```

```python
df2.merge(df1,how='outer')
```

```python
df2.merge(df1,how='right')
```

```python
df2.merge(df1,how='left')
```

```python
df1.merge(df2, on='Name', how = 'outer')
```
